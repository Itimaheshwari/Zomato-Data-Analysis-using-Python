```python
df_country=pd.read_csv("Country-Code.csv")
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
# checking data types
final_df.dtypes
```

```python
country_name=final_df.Country.value_counts().index
country_name
```

```python
country_value=final_df.Country.value_counts().values
country_value
```

```python
# pie chart (top 3 countries that uses zomato)
plt.pie(country_value[:3],labels=country_name[:3],autopct='%1.2f%%')
```

observation: zomato maximum records or transactions are from India after that USA then United kingdoms

```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

observation:
1. when Rating is between 4.5 to 4.9---> Excellent
2. when Rating is between 4.0 to 3.4---> Very Good
3. when Rating is between 3.5 to 3.9---> Good
4. when Rating is between 3.0 to 3.4---> Average
5. when Rating is between 2.5 to 2.9---> Average
6. when Rating is between 2.0 to 2.4---> Poor

```python
import matplotlib
matplotlib.rcParams["figure.figsize"]=(10,4)
sns.barplot(x="Aggregate rating",y= "Rating Count",hue="Rating color", data= ratings,palette=['White','Red','Orange','Yellow','Green','Green'])
```

Observation:
1. Not rated count is very high
2. Maximum numbers of rating  are between 2.5 to 3.4

```python
# count plot
sns.countplot(x="Rating color",data=ratings,palette=['White','Red','Orange','Yellow','Green','Green'])
```

```python
# find out the countries name that has given 0 rating
final_df.groupby(['Aggregate rating','Country']).size().reset_index().head(5)
```
